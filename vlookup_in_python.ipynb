```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
easy_insights = pd.read_csv('easy_insights_data_nov_2022.csv')
easy_insights.head()
```

```python
utm_prod = pd.read_csv('new_utm_data.csv')
utm_prod.head()
```

```python
df1 = easy_insights[['Lead ID','UTM Source','UTM Campaign','UTM Medium','UTM Term','UTM Content']]
df1
```

```python
df2 = utm_prod[['lead_id','campaign_source','campaign_name','campaign_medium','campaign_term','campaign_content']]
df2
```

```python
df = pd.merge(left=df1, right=df2,left_on='Lead ID', right_on='lead_id', how='right')
df
```

```python
import sys
```

```python
df['Lead ID'] = df['Lead ID'].values.astype(str)
df
```

```python
df['lead_id'] = df['lead_id'].values.astype(str)
df
```

```python
df.to_csv('utm_params_check_nov_2022_new222.csv')
```
